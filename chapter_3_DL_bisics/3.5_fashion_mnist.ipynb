# 图像分类数据集（Fashion MNIST）
Fashion-MNIST 是一个10类的服饰分类的数据集。  
torchvision包：服务于pytorch框架 主要组成部分：  
1.torchvision.datasets:一些加载数据的函数以及常用的数据集接口；  
2.torchvision.models:包含常用的模型结构（含预训练模型）（e.g. AlexNet VGG ResNet）  
3.torchvision.transforms:常用的图片变换  e.g 裁剪、旋转等  
4.torchvision.utils：其他的一些有用的方法  

```python
#导入需要的包和模块
import torch
import torchvision
import torchvision.transforms as transforms
import matplotlib.pyplot as plt
import time
import sys
sys.path.append("..")# 为了导入上层目录的d2lzh_pytorch
import d2lzh_pytorch as d2l

mnist_train = torchvision.datasets.FashionMNIST(root = '~/Datasets/FashionMNIST',train=True,download=True,transform=transforms.ToTensor())
mnist_test = torchvision.datasets.FashionMNIST(root='~/Datasets/FashionMNIST',train=False,download=True,transform=transforms.ToTensor())

# print(type(mnist_train))
# print(len(mnist_train),len(mnist_test))
# features,label = mnist_train[0]
# print(features.shape,label)

def get_fashion_mnist_labels(labels):
    text_labels = ['T-shirt','Trouser','Pullover','Dress','Coat','Sandal','Shirt','Sneaker','bag','ankle bookt']
    return [text_labels[int(i)] for i in labels]

# 下面定义一个可以在一行里画出多张图像的对应标签的函数
def show_fashion_mnist(images,labels):
    d2l.use_svg_display()
    _,figs = plt.subplots(1,len(images),figsize=(12,12))
    for f,img, lbl in zip(figs,images,labels):
        f.imshow(img.view((28,28)).numpy())
        f.set_title(lbl)
        f.axes.get_xaxis().set_visible(False)
        f.axes.get_yaxis().set_visible(False)
    plt.show()
    
# 检视训练集中前10个样本的图像内容和文本标签
X,y = [],[]
for i in range(5):
    X.append(mnist_train[i][0])
    y.append(mnist_train[i][1])
show_fashion_mnist(X,get_fashion_mnist_labels(y))
```
